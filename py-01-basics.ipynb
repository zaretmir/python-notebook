# 1. How Python Code is Executed

Los lenguajes de programacin deben compilarse a cdigo mquina. El cdigo mquina es distinto segn el tipo de procesador, por lo que un cdigo compilado para Windows por no va a funcionar en Mac (es el caso de C). Sin embargo, **Python** siguen otro procedimiento.

1. CPython (nuestro compilador) compila nuestro cdigo Python a **Python Bytecode**.
2. Python Bytecode es *platform independent*, por lo que podemos llevarlo a cualquier mquina.
3. Para poder ejecutar ese cdigo en nuestra mquina, necesitamos convertirlo a cdigo mquina. Cuando ejecutamos nuestro cdigo, **Python Virtual Machine** lo convierte en cdigo mquina en tiempo de ejecucin *runtime*.

## Mutable and immutable types

Las variables no son ms que etiquetas para los espacios de memoria en los que se almacena el valor. Podemos obtener la direccin de memoria utilizando la funcin *id()*:

````python
x = 1;
print(id(x))
````

### Primitive types are **immutable**
 Recordemos que los **tipos primitivos son inmutables**. Eso significa que cada vez que reasignamos el valor a una variable, el nuevo valor se almacenar en una nueva direccin de memoria.

 Como dejamos de tener una referencia a la primera direccin de memoria, llegar un momento en el que el Python Garbage Collector libere ese espacio.

````python
x = 12
print(id(x))
x = 15
print(id(x))
````

### Other types are mutable
Eso significa que aunque cambiemos su valor, la direccin de memoria ser la misma.

````python
x = [1]
print(id(x))

x.append(3)
print(id(x))
````

#  Strings

NB: Como los strings son primitive types, y con ello inmutables, cada vez que hacemos un slice, el resultado se almacena en una nueva direccin de memoria.

##  Functions we can apply to strings

````python
course = "Programming with Python"
# length
print(len(course))

# access character
print(course[0])
print(course[-1])

# slicing
print(course[0:3])
print(course[:3])
print(course[4:])
print(course[:])

# los reasultados van a una nueva direccin de memoria
print(id(course))
print(id(course[0:3]))
````

 `range` es un tipo de objeto, y `list` es otro tipo de objeto distinto. Un range ocupa menos que una lista con el mismo nmero de elementos.

# Functions

Las funciones de Python pueden tener:
- Default arguments
- Keyword Arguments
- Podemos usar type hinting, como vimos anteriormente en la seccin de Type Annotation.

 Las **tuplas** son como listas, pero INMUTABLES.

````python
def sum(a, b):
    return a + b
print(sum(2,3))

# con default argument
def sum(a, b=1):
    return a + b
print(sum(2))

# con keyword argument
def sum(a, b):
    return a + b
print(sum(2, b=4))

# con type annotation
def sum(a: int, b:int) -> tuple:
    return (a, a+b)
print(sum(2, 4))
````

## *args

Cuando queremos pasarle a una funcin un nmero indeterminado de parmetros, podemos usar una lista, o podemos usar xargs. Es como `...args` en JavaScript. La notacin es como se ve en el ejemplo (fijarse en el asterisco `*`). Python interpretar el argumento como una **tupla**.

````python
def sumAll(*the_list):
    print(the_list)
````

## **args

Tiene un efecto que recuerda a asignar objetos en Javascript. La idea es utilizar **keyword arguments**, y el resultado ser un **dictionary** (que es el equivalente a un objeto de JS.

````python
def showUser(**user):
    print(user)
    print(user["name"])
    
showUser(name="Hanna", age="28")
````
